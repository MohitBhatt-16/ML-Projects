```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Importing Dataset
diabetes = pd.read_csv('diabetes.csv')
```

```python
diabetes.head()
```

```python
diabetes.shape
```

```python
diabetes.describe()
```

```python
diabetes['Outcome'].value_counts()
```

 0 = Non Diabetic
 1 = Diabetic

# ***Handling Imbalanced Dataset***

```python
non_diabetic = diabetes[diabetes.Outcome==0]
diabetic = diabetes[diabetes.Outcome==1]
```

```python
non_diabetic.shape
```

```python
diabetic.shape
```

```python
non_diabetic_sample = non_diabetic.sample(diabetic.shape[0])
```

```python
non_diabetic_sample.shape
```

```python
diabetic_new = pd.concat([diabetic,non_diabetic_sample], axis=0)
```

```python
diabetic_new.shape
```

```python
diabetic_new.head()
```

```python
diabetic_new['Outcome'].value_counts()
```

```python
#Seperating Data and Labels
X = diabetic_new.drop(columns = 'Outcome', axis = 1)
Y = diabetic_new['Outcome']
```

## ***Data Standardization***

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X = scaler.transform(X)
```

```python
print(X)
```

```python
Y
```

## ***Train Test Split***

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2, stratify=Y)
```

```python
print(X.shape)
```

```python
print(X_train.shape)
```

```python
print(X_test.shape)
```

```python
Y.shape
```

## ***Training Support Vector Machine***

```python
classifier = svm.SVC(kernel = 'linear')
classifier.fit(X_train,y_train)
```

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print(f'Training Data Accuracy {training_data_accuracy}')
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print(f'Test Data Accuracy {test_data_accuracy}')
```
